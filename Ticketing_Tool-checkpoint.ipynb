```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from math import sqrt

def cramers_v(x, y):
    confusion_matrix_data = pd.crosstab(x, y)
    chi2, _, _, _ = chi2_contingency(confusion_matrix_data)
    n = confusion_matrix_data.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix_data.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

# Assuming df is your DataFrame with categorical columns
columns = df.select_dtypes(include='object').columns  # Select only categorical columns
cramers_matrix = pd.DataFrame(index=columns, columns=columns)
cramers_matrix = cramers_matrix.fillna(0).astype(float)

# Compute Cramér's V for each pair of categorical features
for i in range(len(columns)):
    for j in range(i, len(columns)):
        feature1 = columns[i]
        feature2 = columns[j]
        cramers_matrix.loc[feature1, feature2] = cramers_matrix.loc[feature2, feature1] = cramers_v(df[feature1], df[feature2])

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cramers_matrix, annot=True, cmap='coolwarm', fmt=".3f", linewidths=.5)
plt.title("Cramér's V Heatmap for Categorical Features")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Get unique agents in the DataFrame
agents = df['Assigned to'].unique()

# Create separate bar charts for each agent
for agent in agents:
    agent_df = df[df['Assigned to'] == agent]
    
    plt.figure(figsize=(8, 6))
    sns.countplot(x='Corporation', data=agent_df)
    plt.title(f'Tickets Solved by {agent} for Each Corporation Type')
    plt.xlabel('Corporation Type')
    plt.ylabel('Number of Tickets')
    plt.show()
```
