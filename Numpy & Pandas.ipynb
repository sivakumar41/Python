```python
import pandas as pd
```

```python
dataset={
    'cars': ["bmw","volvo","audi"],  #dictionary format and have comma btw every attribute strings should be in quotations only.
    'price':[2,3,4]
         }
print(pd.DataFrame(dataset))
print("\n")

# creation of the series data structure by using various items
li=[2,34,5,456,76]
res=pd.Series(li)   #from the list
print(res)
print("\n")
dic={'a':1,'b':2,'c':3,'d':4,}  # here the  keys will work as the index
print(pd.Series(dic))
print("\n")
x=pd.Series([234,234,234,4],index=[1,1,2, 13])   #here we can give any index the synatx is pd.Series([],index=[])
print(x)
print("\n")

# intdocution to the 2nd data structure data frame
l=["siva","dhan","sravani","maneesha","manasaaaa"]
print(pd.DataFrame(l+l))
print(pd.DataFrame())
print("\n")

# creating the employee dataframe by using the dictionary
d1={
    'name':["siva",'kumar','dhana'],
    'age':[12,32,20],
    'address':['drl','nuz','banglore'],
    'degree':['btech','mca','MTech']
}
res=pd.DataFrame(d1)
print(res)
# now we try to select only some certain cols
print(res[['name','degree']])
res.to_csv("data.csv")
print("the dimension of the dataframe is:",res.ndim,res.shape,res.size)
# working with
print(res)

print("\n")
#  accessing the data with the help of certain functions
print(res.loc[:,'name'])  # one cols alll rows
print(res.loc[:,['name','degree']]) #two cols
print(res.iloc[0:2,[0,1]])   # to get with index
print(res.at[0,'name'])

print("   working with iat ")
print(res.iat[2,1])   # here we will give like matrix
print("\n")
## now we will see the data modifying
res.loc[2,'age']=[121]    # ASSIGNING THE VALUE
print(res.loc[2,'age'])
res.loc[1,'name']=['dhana']
print(res)
```
