```python
import pandas as pd
import numpy as np
```

# Reading and preprocessing the file

```python
df=pd.read_excel("f1.xlsx",keep_default_na=False,header=None)#keep_default_na=False to stop showing the NaN values
df
```

```python
df.drop(df.head(3).index,inplace=True)
```

```python
df.drop(df.tail(1).index,inplace=True)
```

```python
df1=pd.read_csv("Output.csv",keep_default_na=False)
df1
```

```python
#placing the " " with the NaN to work on it
df.replace("",np.nan,inplace=True)
df.shape
```

```python
#Remove the empty columns and rows
df.dropna(how="all",axis=1,inplace=True)
df.dropna(how="all",axis=0,inplace=True)
print(df.shape)
df
```

```python
#After removing the emppty columns and rows we need to adjust the index
df.reset_index(inplace=True,drop=True)
df.columns=np.arange(1,len(df.columns)+1)
df
```

```python
#Here we will drop the rows based on the column value like if particular column dont have value then we drop that row
df.dropna(subset=[df.columns[-1]],inplace=True)
df
```

```python
#removing the rows and columns where we have the minumum NaN
df.dropna(axis="index",inplace=True,thresh=4)
df.dropna(axis="columns",inplace=True,thresh=4)
df.dropna(axis="columns",inplace=True,how="all")
df
```

```python
l=df.iloc[0].to_list()
print(l)
df.columns=l
df
```

```python
#Remove the unwanted string in btw the data like code room amount ect
filter=df.apply(lambda row: row.str.contains("Code",na=False),axis=1)
df=df[~filter]
df.dropna(inplace=True)
df
```

```python
#assigning the correct index to the rows
df.reset_index(inplace=True,drop=True)
```

```python
am_list=df['Amount'].to_list()
print(len(am_list))
    
```

```python
nl=[]
for i in am_list:
    
    if i.startswith("(") and i.endswith(")"):
        i=i.replace("("," ").replace(")"," ")
        nl.append(i)
    nl.append(i)

        
```

```python
def remove_parentheses(amount):
    return amount.replace('(', '').replace(')', '')
df['Amount'] = df['Amount'].apply(remove_parentheses)
df.shape
```
